```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
import pandas as pd
```

```python
df=pd.read_excel(r'E:\CAPSTONE\FILES TO BE UPLOADED\DatsetUsedForTrainingLDA.xlsx')
```

```python
Skill_Null=df[df['Skills']=='[]']
```

```python
df_mod=df.copy()
```

```python
def skillKeyword(skil,summ):
    skill=skil
    summary=summ
    if skill=='[]':
        try:
            skill_key=str(keywords(summary,words=3,lemmatize=True,pos_filter='NNPS').split('\n'))
            return skill_key
        except:
            return skill
    else:
        return skill
```

```python
df_mod['New_Skills']=df_mod.apply(lambda row:skillKeyword(row['Skills'],row['Summary_Final']),axis=1)
```
